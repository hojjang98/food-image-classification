```python
# Cell 1: í¬ë¡¤ë§ í•¨ìˆ˜ ì •ì˜
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager

import os, time, urllib

def fetch_google_images(query, max_images, save_dir):
    # í¬ë¡¬ ì˜µì…˜ ì„¤ì •
    options = webdriver.ChromeOptions()
    options.add_argument('--no-sandbox')
    options.add_argument('--disable-dev-shm-usage')
    options.add_argument(
        "user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 "
        "(KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
    )

    # ë“œë¼ì´ë²„ ì‹¤í–‰
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
    driver.get(f"https://www.google.com/search?q={query}&tbm=isch")
    time.sleep(2)

    # ìŠ¤í¬ë¡¤ ë°˜ë³µ
    last_height = driver.execute_script("return document.body.scrollHeight")
    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height

    # ì´ë¯¸ì§€ ë‹¤ìš´ë¡œë“œ
    os.makedirs(save_dir, exist_ok=True)
    existing = len(os.listdir(save_dir)) if os.path.exists(save_dir) else 0
    count = existing

    imgs = driver.find_elements(By.TAG_NAME, "img")
    print(f"[ðŸ”Ž] Found {len(imgs)} image tags for '{query}'")

    for img in imgs:
        try:
            src = img.get_attribute("src")
            if src and "http" in src:
                save_path = os.path.join(save_dir, f"{query.replace(' ', '_')}_{count}.jpg")
                urllib.request.urlretrieve(src, save_path)
                print(f"  [{count}] Saved: {src}")
                count += 1
                if count >= existing + max_images:
                    break
        except Exception:
            continue

    driver.quit()
    print(f"[âœ…] {count - existing} new images saved for '{query}'")
```

```python
# Cell 2: í´ëž˜ìŠ¤ë³„ í¬ë¡¤ë§ ì‹¤í–‰
classes = [
    "apple", "banana", "beetroot", "bell pepper", "cabbage", "capsicum",
    "carrot", "cauliflower", "chilli pepper", "corn", "cucumber", "eggplant",
    "garlic", "ginger", "grapes", "jalapeno", "kiwi", "lemon", "lettuce", "mango",
    "onion", "orange", "papaya", "passion fruit", "pear", "peas", "pineapple",
    "pomegranate", "potato", "raddish", "soy beans", "spinach", "sweet potato",
    "tomato", "turnip", "watermelon"
]

# í¬ë¡¤ë§ ì‹¤í–‰
for cls in classes:
    search_query = f"raw {cls}" if cls not in ["grapes", "kiwi", "orange", "apple"] else f"whole {cls}"
    print(f"\nðŸ”„ Crawling images for: {search_query}")
    fetch_google_images(
        query=search_query,
        max_images=100,
        save_dir=f"./images/{cls}"
    )
```

```python
# Cell 3: ì´ë¯¸ì§€ ì •ì œ í•¨ìˆ˜
import cv2

def clean_images(folder_path, min_width=100, min_height=100):
    for fname in os.listdir(folder_path):
        fpath = os.path.join(folder_path, fname)
        try:
            img = cv2.imread(fpath)
            if img is None:
                print(f"ðŸ—‘ï¸ Removing corrupted: {fname}")
                os.remove(fpath)
            elif img.shape[0] < min_height or img.shape[1] < min_width:
                print(f"ðŸ—‘ï¸ Removing small image: {fname}")
                os.remove(fpath)
        except Exception as e:
            print(f"ðŸ—‘ï¸ Error reading {fname}: {e}")
            os.remove(fpath)

# ì „ì²´ í´ëž˜ìŠ¤ì— ëŒ€í•´ ì •ì œ ì‹¤í–‰
for cls in classes:
    print(f"\nðŸ§¹ Cleaning folder: {cls}")
    clean_images(f"./images/{cls}")
```
